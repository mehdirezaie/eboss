```python
def plot_auto(files, l, c, ls, mk):
    plt.figure(figsize=(12,8))

    #     plt.rc('axes.spines', right=False, top=False)
    # for i in range(20):
    #     b = nn['t']
    #     plt.plot(nn['t'], b*nn['wjacks'][i], 'g', alpha=0.2)
    #     plt.plot(lin['t'], b*lin['wjacks'][i], 'r', alpha=0.2)
    #     plt.plot(uni['t'], b*uni['wjacks'][i], 'b', alpha=0.2)
    #c  = ['blue', 'red', 'green', 'lightblue', 'violet', 'lightgreen']
    #l  = ['Uncorrected', 'Linear weight', 'NN weight', None, None, None]
    for i,f in enumerate(files):
        s = np.load(f).item()
        b = s['t'][:-1]
#         print(s.keys())
        plt.errorbar(b, s['w']-s['dmean']**2, 
                 yerr=s['werr'], ls=ls[i], color=c[i],
                label=l[i], marker=mk[i])
    #plt.plot([1, 10],[0,0], 'k:')
    plt.legend(frameon=False, title='Ngal x Ngal')
    plt.axis()
    # plt.yscale('log')
    plt.ylabel(r'$ \omega$', fontsize=30)
    plt.xlabel(r'$\theta[deg]$', fontsize=30)
    plt.savefig('./figs/xi-ngal.pdf', bbox_inches='tight')
plot_auto(files, ['Uncorrected', 'NN weight','Linear weight'], ['b','g','r'], ['None', '--','-'], 
         ['.', '+', 'x'])
```

```python
def plot_cross():
    bands = ['r','g','z']
    mk = ['.', '+', 'x']
    ls = ['None', '--', '-']
    labels = ['depth-'+b for b in bands]
    labels += ['seeing-'+b for b in bands]
    labels += ['airmass-'+b for b in bands]
    labels += ['EBV','nstar']
    plt.rc('font', size=15, family='serif')
    plt.rc('axes.spines', right=False, top=False)
    f,a = plt.subplots(ncols=3, nrows=4, figsize=(15, 20))
    a   = a.flatten()
    f.delaxes(a[-1])
    # plt.suptitle("eBOSS ELG DECaLS DR5", fontsize=18)
    plt.subplots_adjust(wspace=0.05, hspace=0.2, top=0.97)    
    pre = '/global/cscratch1/sd/mehdi/dr5_anand/eboss/clustering/xi-ebossdr5-cross-ngal'
    file = lambda sys, w: pre+sys+'-'+w+'_nside_256_njack_20.npy'
    c    = ['b','r','g']
    sae  = [[], [], []]
    for i,l in enumerate(labels):
        for j,w in enumerate(['uni','lin', 'nn']):
            s = np.load(file(l, w)).item()
            b = s['t'][:-1]
#             print(s.keys())
            a[i].errorbar(b+j*0.1, s['w']-s['dmean1']*s['dmean2'], yerr=s['werr'], 
                          ls=ls[j], color=c[j], marker=mk[j])
            rse = np.sum(np.abs(s['w']-s['dmean1']*s['dmean2']))#/s['werr']**2)
            #a[i].text(0.1, 0.1+j*0.05, r'SAE/dof={:.3f}/{}'.format(rse, s['werr'].size), 
            #          transform=a[i].transAxes,
            #         color=c[j])
            #sae[j].append(rse)

#             if j == 1:
        if i in [0,3,6,9]:a[i].set_ylabel(r'$\omega$')
        a[i].set_xlabel(r'$\theta$')
        a[i].text(0.5, 0.95, 'Ngal x '+l, fontsize=12, transform=a[i].transAxes)
        if i in [0,3,6,9]:
            a[i].set_ylabel(r'$\omega$')
        else:
            a[i].set_yticks([])
            
        if i > 8:
            a[i].set_ylim(-0.045, 0.01)
        elif i < 3:
            a[i].set_ylim(-0.0005, 0.0005)
        else:
            a[i].set_ylim(-0.005, 0.005)
        a[i].plot([1, 10], [0,0], 'k:')
        #a[i].yaxis.set_major_formatter(FormatStrFormatter('%.e'))
        
    for i,(l,m) in enumerate(zip(['Uncorrected', 'NN weight','Linear weight'], ['b','g','r'])):
        a[0].text(0.5, 0.05+i*0.1, l, color=m, transform=a[0].transAxes)
    #return sae
    
#
#sae = plot_cross()
plot_cross()
plt.savefig('./figs/eboss-dr5-ngalsys.pdf', bbox_inches='tight')
```

## 3. Auto-correlation with upgraded maps of ELG
Here we use the NN answer for nside 256, but we upgrade it, along with the ELG hpmap to nside of 1024. We only run the `xi-jack.py` for the auto-correlation.

```python
from glob import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
files = glob('/global/cscratch1/sd/mehdi/dr5_anand/eboss/clustering/*auto*1024*.npy')
files.sort()
files
```

```python
import sys
sys.path.append('/global/homes/m/mehdi/github/DESILSS')
from cl import cl2xi
```

```python
clt = np.loadtxt('/global/cscratch1/sd/mehdi/dr5_anand/eboss/clustering/ebossELG.dat', usecols=(0,1))
```

```python
cost = np.cos(np.deg2rad(np.arange(0.172*1.5, 10, 0.172)))
xit  = cl2xi(clt[:,1], cost)
```

```python
plt.rc('font', family='serif', size=20)
plt.rc('axes.spines', right=False, top=False)
```

```python
def plot_auto(files, l, c):
    plt.figure(figsize=(12,8))
    plt.rc('font', size=20, family='serif')
    #     plt.rc('axes.spines', right=False, top=False)
    # for i in range(20):
    #     b = nn['t']
    #     plt.plot(nn['t'], b*nn['wjacks'][i], 'g', alpha=0.2)
    #     plt.plot(lin['t'], b*lin['wjacks'][i], 'r', alpha=0.2)
    #     plt.plot(uni['t'], b*uni['wjacks'][i], 'b', alpha=0.2)
    #c  = ['blue', 'red', 'green', 'lightblue', 'violet', 'lightgreen']
    #l  = ['Uncorrected', 'Linear weight', 'NN weight', None, None, None]
    for i,f in enumerate(files):
        s = np.load(f).item()
        b = s['t']
        plt.errorbar(b, b*(s['w']-s['dmean']**2), 
                 yerr=b*s['werr'], ls='--', color=c[i],
                label=l[i])
    #plt.plot([1, 10],[0,0], 'k:')
    t = np.rad2deg(np.arccos(cost))
    plt.plot(t, 5*t*xit, 'k-', label="5xMartin's")
    plt.legend(frameon=False, title='Ngal x Ngal')
    plt.axis()
#     plt.loglog()
#     plt.yscale('log')
    plt.ylabel(r'$\theta x \omega$')
    plt.xlabel(r'$\theta[deg]$')
    # plt.savefig('./figs/xi-ngal.pdf', bbox_inches='tight')
plot_auto(files, ['Linear weight', 'NN weight', 'Uncorrected'], ['r','g','b'])
plt.savefig('./figs/xi-dr5-theory256.pdf')
```
