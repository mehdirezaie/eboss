````python
%matplotlib inline
import matplotlib.pyplot as plt
import sys
sys.path.append('/home/mehdi/github/LSSutils')
from LSSutils import dataviz
from glob import glob
````

````python
files = glob('/home/mehdi/data/eboss/v6/results_sgc.all/ablation_tol/*.npy')
files
````

````python
def plot(versus='MSEall'):
    fig, ax = plt.subplots(ncols=3, nrows=3, figsize=(15, 15))
    ax = ax.flatten()

    for i,file_i in enumerate(files):
        name = file_i.split('/')[-1].split('_fold0')[0]
        dataviz.ablation_plot(file_i,
                              labels=None,
                              allow_pickle=True,
                              annot=False,
                              hold=True,
                              saveto=None,
                              ax=ax[i], 
                              versus=versus)
        ax[i].set_title(name)
    plt.show()
````

Relative to the model with all features (`baseline')

````python
plot('MSEall')
````

Relative to a constant model (the simple model)

````python
plot('baselineMSE')
````
