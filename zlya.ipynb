## plot

```python
def add_pk(pk, ax, **kw):
    ax.plot(pk.poles['k'], pk.poles['power_0'].real-pk.attrs['shotnoise'], **kw)
    return ax

fig, ax = plt.subplots()

for cap in ['ngc', 'sgc']:
    
    ax = add_pk(pks[f'{cap}_zpca'], ax, label=f'{cap} PCA')
    ax = add_pk(pks[f'{cap}_zlya'], ax, label=f'{cap} Lya')
    
ax.loglog()
ax.legend()
ax.set_ylabel('P0')
```

```python
def add_pk(pk, ax, ell=0, **kw):
    if ell==0:
        ax.plot(pk.poles['k'], pk.poles[f'power_{ell}'].real-pk.attrs['shotnoise'], **kw)
    else:
        ax.plot(pk.poles['k'], pk.poles[f'power_{ell}'].real, **kw)
    return ax

fig, ax = plt.subplots()


ELL = 2

for cap in ['ngc', 'sgc']:
    
    
    ax = add_pk(pks[f'{cap}_zpca_highz'], ax, ell=ELL, label=f'{cap} PCA')
    ax = add_pk(pks[f'{cap}_zlya_highz'], ax, ell=ELL, label=f'{cap} Lya')
    
#ax.loglog()
ax.set_xscale('log')
ax.legend()
ax.set_ylabel(f'$P_{ELL}$')
```
